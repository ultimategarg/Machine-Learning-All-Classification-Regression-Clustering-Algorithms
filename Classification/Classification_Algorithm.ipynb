# Logistic Regression:

```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Data/Social_Network_Ads.csv")

#divide the dataset into features and Labels
features=df.iloc[:,[2,3]].values
labels=df.iloc[:,4].values
```

```python
#Spliting the dataset into training and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.20,random_state=0)
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
a=LogisticRegression(random_state=0)

#Fitting the Logistic Regression to training set
a.fit(x_train,y_train)
pred=a.predict(x_test)
```

```python
#Confusion Matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
print(cm)
```

```python
#Visualising the training set Results
from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,a.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set== j, 0], x_set[y_set== j, 1],alpha=0.8,c=ListedColormap(('red','green'))(i), label=j)
plt.xlabel("age")
plt.legend()   
plt.show()
    
```

```python
#Visualising the test set Results
from matplotlib.colors import ListedColormap
x_set,y_set=x_test,y_test
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,a.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set== j, 0], x_set[y_set== j, 1],alpha=0.8,c=ListedColormap(('red','green'))(i), label=j)
plt.xlabel("age")
plt.legend()   
plt.show()
```

# K-Nearest Neighbors:

```python
#import the libraries
import pandas as pd
df=pd.read_csv("Data/Social_Network_Ads.csv")

```

```python
features=df.iloc[:,[2,3]].values
labels=df.iloc[:,4].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python

from sklearn.neighbors import KNeighborsClassifier
a= KNeighborsClassifier(n_neighbors=5,p=2)

#Fitting the KNN to training set
a.fit(x_train,y_train)
pred=a.predict(x_test)


from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
#Visualising the training set Results
from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,a.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set== j, 0], x_set[y_set== j, 1],c=ListedColormap(('red','green'))(i), label=j)
plt.xlabel("age")
plt.legend()   
plt.show()
```
