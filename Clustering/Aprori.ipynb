# Aprori:

```python
#import the libraries
import pandas as pd
from apyori import apriori
```

```python
#Each row Represents a single Customer
#Data of a Weak's Transaction in a store
data=pd.read_csv("Data/Market_Basket_Optimisation.csv",header=None)
data.head()
```

```python
#transaction as t
t=[]
for i in range(5):
    t.append([str(data.values[i,j]) for j in range (10)])
    
```

```python
#training Apriori on the data
rules=apriori(t,min_support=0.003,min_confidence=0.2,min_left=3,min_lenght=2)
```

```python
#visualing the Results
list(rules)
```

```python
#Example
transactions = [
    ['beer', 'nuts'],
    ['beer', 'cheese'],
]
results = list(apriori(transactions))
results
```
